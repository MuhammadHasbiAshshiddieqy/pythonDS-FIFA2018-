```python
data_match["Team"] = data_match["Team"].str.upper()
data_cou["Country"] = data_cou["Country"].str.upper()
new_feature = data_match.merge(data_cou, left_on="Team", right_on="Country", how="left")
```

Import additional package such as **Matplotlib** and **Seaborn**. 

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data_match.head()
```

**1. Heatmap**

Pembuatan heatmap bertujuan untuk mengetahui hubungan pengaruh antar atribut pada data yang telah tertampung pada variabel **data_match**. Nilai dari heatmap menunjukan bahwa semakin tinggi nilai pada relasi atribut, maka semakin tinggi keterkaitan antar atribut tersebut. Dan ketika relasi atribut pada heatmap memiliki nilai kurang dari 0, maka relasi antar atribut bersangkutan adalah saling berbanding terbalik.

```python
plt.figure(figsize=(20,20))
sns.heatmap(data_match.corr(), annot=True)
plt.show()
```

**2. Buble Chart**

Pembuatan Buble Chart **The Influence of Passes on Ball Possession** bertujuan untuk mengetahui pengaruh **Passing** terhadap **Ball Possession**. Dari Buble Chart tersebut dapat disimpulkan bahwa semakin sering **Passing**, maka akan meningkatkan **Ball Possession**. 

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x="Passes", y="Ball Possession %", size="Passes", hue="Passes", data=data_match)
plt.title("The Influence of Passes on Ball Possession", size=20)
plt.xlabel("Passes")
plt.ylabel("Ball Possession (%)")
plt.legend()
plt.show()
```

**3. Bar Chart**

Bar Chart **Goal Scored in Europe Continent** digunakan untuk membandingkan jumlah goal dari tiap negara pada kontinen Eropa.

```python
europe_goal = pd.DataFrame(new_feature[new_feature["Continent"]=="Europe"].groupby("Team")["Goal Scored"].sum())
europe_goal.head()
```

```python
europe_goal.plot(kind="barh", figsize=(10,8))
plt.title("Goal Scored in Europe Continent", size=20)
plt.ylabel("Country")
plt.xlabel("Goal Scored")
plt.show()
```

**4. Box Plot**

Box Plot **Fouls Commited (Europe and South America)** bertujuan untuk mengetahui persebaran **Fouls Commited** pada kontinen Eropa dan Amerika.

```python
europe_and_america_fouls = pd.DataFrame(new_feature[(new_feature["Continent"]=="Europe") | (new_feature["Continent"]=="South America")].groupby(["Continent","Team"])["Fouls Committed"].sum())
europe_and_america_fouls = europe_and_america_fouls.reset_index()
europe_and_america_fouls.head()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x = europe_and_america_fouls["Continent"], y = europe_and_america_fouls["Fouls Committed"])
plt.title("Fouls Commited (Europe and South America)", size=20)
plt.show()
```
